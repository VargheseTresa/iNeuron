### 1. Define a class with a generator which can iterate the numbers, which are divisible by 7, between a given range 0 and n.

```python
class NumGenerator:
    
    def generateSeries7(self, n):
        for i in range(n):
            if i%7 == 0:
                yield i
                
                
obj = NumGenerator()
n = 2500

print("Numbers divisible by 7 upto ", n)
for num in obj.generateSeries7(n):
    print(num)
```

### 2. Write a program to compute the frequency of the words from the input. The output should output after sorting the key alphanumerically.

```python
inpString = input("Enter the sequence : ").split()

wordCount = {}

for word in inpString:
    if word in wordCount:
        wordCount[word] += 1
    else:
        wordCount[word] = 1
        
for key in sorted(wordCount.keys()):
    print(key, ":", wordCount[key])
```

### 4. Please write a program to generate all sentences where subject is in [&quot;I&quot;, &quot;You&quot;] and verb is in [&quot;Play&quot;, &quot;Love&quot;] and the object is in [&quot;Hockey&quot;,&quot;Football&quot;].

```python
subjects = ["I", "You"]
verbs = ["Play", "Love"]
objects = ["Hockey","Football"]
```

```python
for subj in subjects:
    for verb in verbs:
        for obj in objects:
            print(subj, verb, obj)
```

### 5. Please write a program to compress and decompress the string &quot;hello world!helloworld!hello world!hello world!&quot;.

```python
import zlib

strInput = "hello world!helloworld!hello world!hello world!"
compressed = zlib.compress(bytes(strInput, 'utf-8'))
print("After compressing : ", compressed)

print("After decompressing : ", zlib.decompress(compressed))
```

### 6. Please write a binary search function which searches an item in a sorted list. The function should return the index of element to be searched in the list.

```python
def binarySearch(arr, minIndex, maxIndex, x):
    
    if maxIndex >= minIndex:
        
        midIndex = (minIndex + maxIndex) // 2
        
        if arr[midIndex] == x:
            return midIndex
        elif arr[midIndex] > x:
            return binarySearch(arr, minIndex, midIndex-1, x)
        else:
            return binarySearch(arr, midIndex + 1, maxIndex, x)
    else:
        return -1 # element not found
        



inputList = [3,5,2,4,8,11,9]
x = 3

print("Sorted array : ", sorted(inputList))
print("Position of element ", x)
binarySearch(sorted(inputList), 0, len(inputList)-1, x)
```
